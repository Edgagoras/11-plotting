```python
# explicit import are preferred
import bokeh
import bokeh.io
import bokeh.plotting
import pandas as pd
import numpy as np

# hide warnings
import warnings
warnings.filterwarnings("ignore", category=RuntimeWarning)
```

```python
bokeh.io.output_notebook()
```

## Prepare some data

```python
# prepare some data
N = 100
x = np.linspace(0, 4 * np.pi, N)
y0 = np.sin(x)
y1 = np.cos(x)
y2 = np.sin(x) + np.cos(x)
```

### Create individual plots

```python
# create a new plot
p1 = bokeh.plotting.figure(
    width=250, 
    height=250, 
    title=None)
p1.circle(
    x, y0, 
    size=10, 
    color="navy", 
    alpha=0.5)
```

```python
# create a new plot and share both ranges
p2 = bokeh.plotting.figure(
    width=200, 
    height=250, 
    x_range=p1.x_range, 
    y_range=p1.y_range, 
    title=None)
p2.triangle(
    x, y1, 
    size=10, 
    color="firebrick", 
    alpha=0.5)
```

```python
# create a new plot and share only one range
p3 = bokeh.plotting.figure(
    width=250, 
    height=250, 
    x_range=p1.x_range, 
    title=None)
p3.square(
    x, y2, 
    size=10, 
    color="olive",
    alpha=0.5)
```

### Composite plot

```python
p = bokeh.layouts.gridplot((p1, p2, p3), toolbar_location=None)
bokeh.io.show(p)
                           
```
